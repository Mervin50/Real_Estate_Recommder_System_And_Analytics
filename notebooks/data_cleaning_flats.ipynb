```python
df.head()
```

```python
def treat_price(x):
    if isinstance(x, float) or pd.isna(x):
        return None
    x = x.split(' ')  # move split here inside function
    if len(x) < 2:
        return None  # not enough parts to parse
    try:
        num = float(x[0])
    except ValueError:
        return None

    if x[1].lower() == 'lac':
        return round(num / 100, 2)
    elif x[1].lower() == 'crore':
        return round(num, 2)
    else:
        return None  # unknown unit
```

```python
# Now apply it cleanly
df['price'] = df['price'].apply(treat_price)
```

```python
df.head(5)
```

```python
# price_per_sqft
df['price_per_sqft'].value_counts()
```

```python
def clean_price_per_sqft(x):
    if pd.isna(x):
        return None
    x = x.split('/')[0]  # take left side of '/'
    x = x.replace('₹', '').replace(',', '').strip()
    if x.replace('.', '').isdigit():  # check if it's a number
        return float(x)
    else:
        return None

df['price_per_sqft'] = df['price_per_sqft'].apply(clean_price_per_sqft)
```

```python
# bedrooms
df['bedRoom'].value_counts()
```

```python
df[df['bedRoom'].isnull()]
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
def clean_bedroom(x):
    if pd.isna(x):
        return None
    parts = x.split(' ')
    if parts[0].isdigit():
        return int(parts[0])
    else:
        return None

df['bedRoom'] = df['bedRoom'].apply(clean_bedroom)
```

```python
# bathroom
df['bathroom'].value_counts()
```

```python
df['bathroom'].isnull().sum()
```

```python
def clean_bathroom(x):
    # If the value is NaN, return None
    if pd.isna(x):
        return None
    # Split the string by spaces and get the first part (assuming it's a number)
    parts = x.split(' ')
    if parts[0].isdigit():
        return int(parts[0])  # Convert to integer if it's a number
    else:
        return None  # Return None if it's not a valid number

# Apply the cleaning function to the 'bathroom' column
df['bathroom'] = df['bathroom'].apply(clean_bathroom)
```
